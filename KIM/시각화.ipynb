# Adversarial attack 가시화

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 준비
data = {
    'Model': ['Original', 'BIM', 'FGSM', 'PGD', 'DeepFool'],
    'Loss': [
        0.9779453873634338,  # Original
        3.620353937149048,   # BIM
        6.029921531677246,   # FGSM
        4.351079940795898,   # PGD
        float('nan')         # DeepFool (Loss is NaN)
    ],
    'Accuracy': [
        0.9002624750137329,  # Original
        0.9002624750137329,  # BIM
        0.7454068064689636,  # FGSM
        0.8635170459747314,  # PGD
        0.29921260476112366  # DeepFool
    ]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 그래프 크기 설정
plt.figure(figsize=(14, 7))

# 손실 그래프 생성
plt.subplot(1, 2, 1)
plt.bar(df['Model'], df['Loss'], color='red', alpha=0.7)
plt.title('Loss Comparison Across Different Adversarial Attacks')
plt.xlabel('Model')
plt.ylabel('Loss')
plt.ylim(0, 7)  # 손실 값의 범위를 설정 (DeepFool의 NaN 제외, 최대 값 기준으로 설정)

# 정확도 그래프 생성
plt.subplot(1, 2, 2)
plt.bar(df['Model'], df['Accuracy'], color='blue', alpha=0.7)
plt.title('Accuracy Comparison Across Different Adversarial Attacks')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

# 그래프 간의 간격 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
labels = ['Original', 'BIM', 'FGSM', 'PGD', 'DeepFool']

# 정확도 데이터
accuracy = [
    0.9003,  # Original
    0.9003,  # BIM
    0.7454,  # FGSM
    0.8635,  # PGD
    0.2992   # DeepFool
]

# 손실 데이터 (NaN 값을 0으로 대체)
loss = [
    0.9779,  # Original
    3.6204,  # BIM
    6.0299,  # FGSM
    4.3511,  # PGD
    0.0      # DeepFool (NaN을 0으로 대체)
]

# 파스텔 톤의 색상
colors_pastel = ['#ff9999', '#ffd1b2', '#ffe699', '#c6e2ff', '#d9d9d9']

# 원형 그래프 그리기 - 정확도
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.pie(accuracy, labels=labels, autopct='%1.1f%%', colors=colors_pastel)
plt.title('Accuracy Comparison Across Different Adversarial Attacks')

# 원형 그래프 그리기 - 손실
plt.subplot(1, 2, 2)
plt.pie(loss, labels=labels, autopct='%1.1f%%', colors=colors_pastel)
plt.title('Loss Comparison Across Different Adversarial Attacks')

# 그래프 표시
plt.show()
```
